Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #to train model
from sklearn.linear_model import LogisticRegression  # regression to train model
from sklearn.metrics import accuracy_score # to score accuracy
```

```python
#loading credit card data to a panda dataframe
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

This dataset is highly unbalanced because there 99% points which represent legit ones and only 1% accounts for fraud ones so our model if we train on this dataset wont be able to recognize fraud ones because of fewer availability of fruad data points

0: legit transaction and
1: fraud transaction


```python
# separate legit and fraud data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
#statistics measures of data
legit.Amount.describe()
# std is standard deviation
# % is percentile
```

```python
fraud.Amount.describe()
```

```python
#compare values obtained for both legit and fraud ones
credit_card_data.groupby('Class').mean()
```

Under Sampling: build a sample datset having similar distribution as of legit data and fraud ones in order to remove unbalanced data


we are going to take random 492 values of legit data so that we get equal no of points for both legit and fraud data
and then we join it

```python
legit_randomSample=legit.sample(n=492)
```

```python
new_dataFrame=pd.concat([legit_randomSample,fraud],axis=0) #axis =0 means row and axis=1 is column
```

```python
new_dataFrame.head()
```

```python
new_dataFrame['Class'].value_counts()
```

```python
new_dataFrame.groupby('Class').mean()
```

```python
#splitting data into features and targets
X=new_dataFrame.drop(columns='Class',axis=1)
Y=new_dataFrame['Class']
print(X)
```

```python
print(Y)
```

Split the data into training data and testing data


stratify to evenly distribute Class 0 and Class 1 for training as well test data where 20% of data is test data and 80% is training data

```python
X_train, X_test, Y_train , Y_test =train_test_split(X,Y,test_size=0.2,stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model training : using Logistic training which is best for binary classification data

```python
model=LogisticRegression()
```

```python
#training the logistic regressin model with training data
model.fit(X_train, Y_train)
```

Model Evaluation and Accuracy Score

```python
#accuracy on training data
X_train_predicted=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_predicted, Y_train )
print('Accuracy on training data:', training_data_accuracy)
```

```python
#accuracy of test data
X_test_predicted=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_predicted,Y_test)
print('Accuracy on test data', test_data_accuracy)
```
